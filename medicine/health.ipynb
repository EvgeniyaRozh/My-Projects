### Таблица 005. Распространенность злокачественных новообразований в России в 2011-2021 гг.

```python
# загрузка таблицы
table_005 = pd.read_excel('2021_Таблица_005_Состояние_онко_помощи_в_РФ.xlsx', header=[0, 1, 2, 3])
table_005
```

Удалим последнюю строку с ремаркой. Заменим название столбца '2021*' на '2021'. Приведем таблицу в удобный для анализа вид. Соединим наименование локализации с кодом.

```python
table_005 = table_005.droplevel(0, axis=1) #удаление верхнего уровня индекса
table_005 = table_005.droplevel(0, axis=1)
table_005 = table_005.droplevel(0, axis=1)
table_005.rename(columns = {'Unnamed: 0_level_3':'Локализация, нозологическая форма',\
                            'Unnamed: 1_level_3':'Код МКБ 10', \
                            '2021*' : '2021'}, inplace = True )
# удаление ремарки
table_005 = table_005.drop(index = 25)
table_005 = table_005.reset_index(drop=True)
table_005
```

```python
# составление словаря- код:локализация
dict_005 = pd.Series(table_005['Код МКБ 10'].values,index=table_005['Локализация, нозологическая форма']).to_dict()
dict_005
```

```python
# переворот таблицы
table_005 = table_005.set_index('Локализация, нозологическая форма') #установление индекса
table_005 = table_005.stack().reset_index()

# переимеование столбца
table_005.rename(columns = {'level_1':'Год', 0:'Численность контингента больных на 100 тыс. населения'}, inplace = True)

table_005
```

```python
# удаление лишних строк
table_005 = table_005[table_005['Год'] != 'Код МКБ 10']
table_005
```

```python
# вставка столбца с кодом заболевания
table_005["Код МКБ 10"] = table_005["Локализация, нозологическая форма"].map(dict_005) 

# замена значения столбца, чтоб было в едином стиле для удобного дальнейшего объединения
table_005['Локализация, нозологическая форма'] =  table_005['Локализация, нозологическая форма']\
                                                   .str.replace('Все злокачествен. новообразования',\
                                                                'Все злокачественные новообразования')
table_005['Код МКБ 10'] =  table_005['Код МКБ 10'].str.replace('С81-96','C81-96')

table_005
```

```python
table_005.info()
```

Таблица состоит из 275 строк и 4 столбцов. Название колонок удобночитаемы. Пропущенных значений нет. Данные по численности переведем в вещественный тип.
  
Описание данных  
В каждой строке содержатся количество заболеваний по локализации за каждый год.

```python
# перевод типа данных
table_005['Численность контингента больных на 100 тыс. населения'] =table_005['Численность контингента больных на 100 тыс. населения'].astype(float)
table_005['Год'] = table_005['Год'].astype(int)
```
